```python
import pandas as pd
```

```python
df = pd.read_csv('C:/Users/Oscar Florez/OneDrive/Documents/GitHub/Clase4_Data/Scripts/police_deaths_USA_v6.csv')
df.head()
```

```python
# Agrupar por 'State' y contar el n√∫mero de asesinatos
state_counts = df.groupby('State').size().reset_index(name='Count')

# Mostrar el reporte
print(state_counts)
```

```python
# Ordenar por 'Count' en orden descendente
state_counts_sorted = state_counts.sort_values(by='Count', ascending=False)

# Seleccionar las 10 ciudades con m√°s asesinatos
top_10_states = state_counts_sorted.head(10)

# Agrupar el resto bajo "Otros"
others = pd.DataFrame([['Otros', state_counts_sorted.iloc[10:]['Count'].sum()]], columns=['State', 'Count'])

# Concatenar las 10 ciudades con "Otros"
final_df = pd.concat([top_10_states, others], ignore_index=True)

# A√±adir una fila con el total
total = pd.DataFrame([['Total', final_df['Count'].sum()]], columns=['State', 'Count'])
final_df = pd.concat([final_df, total], ignore_index=True)

# Mostrar el resultado
print(final_df)
```

```python
# Agrupar por 'Cause' y contar el n√∫mero de ocurrencias
cause_counts = df.groupby('Cause').size().reset_index(name='Count')

# Mostrar el resultado
print(cause_counts)
```

```python
# Ordenar por 'Count' en orden descendente
cause_counts_sorted = cause_counts.sort_values(by='Count', ascending=False)

# Seleccionar las 10 causas con m√°s ocurrencias
top_10_causes = cause_counts_sorted.head(10)

# Agrupar el resto bajo "Otros"
others = pd.DataFrame([['Otros', cause_counts_sorted.iloc[10:]['Count'].sum()]], columns=['Cause', 'Count'])

# Concatenar las 10 causas con "Otros"
final_cause_df = pd.concat([top_10_causes, others], ignore_index=True)

# A√±adir una fila con el total
total = pd.DataFrame([['Total', final_cause_df['Count'].sum()]], columns=['Cause', 'Count'])
final_cause_df = pd.concat([final_cause_df, total], ignore_index=True)

# Mostrar el resultado
print(final_cause_df)
```
